```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

census_walk_scores = pd.read_csv('CSVs/census_walkscore_by_zip.csv')
sorted_cen_scores = census_walk_scores.sort_values('Walk Score', ascending=False).reset_index(drop=True)

zip_codes = sorted_cen_scores['Zipcode']
walk_scores = round(sorted_cen_scores['Walk Score'], 2)

med_inc_12 = sorted_cen_scores['Median Income 2012']
med_inc_18 = sorted_cen_scores['Median Income 2018']

zip_code = [str(i) for i in zip_codes]
walk_score = [str(i) for i in walk_scores]

zip_scores = zip(zip_code, walk_score)

width = 0.35
x = np.arange(len(zip_codes))

fig, ax = plt.subplots(figsize=(15,10))
bars1 = ax.bar(x - width/2, med_inc_12, width, label='2012')
bars2 = ax.bar(x + width/2, med_inc_18, width, label='2018')

ax.set_ylabel('Median Income', fontsize=14.0)
ax.set_title('Median Income 2012 vs 2018', fontsize=16.0)
ax.set_xticks(x)
ax.set_xlabel('Zip Codes/Walk Scores', fontsize=14.0, labelpad=20)
ax.set_xticklabels(zip_scores, fontsize=12.0)
ax.tick_params('x', labelrotation=77)
ax.legend(title='Key', title_fontsize=12.0, loc='best', shadow=True, fontsize=12.0)

fig.tight_layout()
plt.show()
```
