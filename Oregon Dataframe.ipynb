```python
import pandas as pd
from config import gkey
import requests
import json


maps_base_url = 'https://maps.googleapis.com/maps/api/geocode/json?'
```

```python
oregon_df = pd.read_csv(r'OR_Features_20200701.txt', sep='|')
oregon_df
```

```python
multnomah_df = oregon_df.loc[oregon_df['COUNTY_NAME'] == 'Multnomah', :]
```

```python
multnomah_feature_df = multnomah_df.loc[(multnomah_df['FEATURE_CLASS'] == 'Populated Place') 
                                        | (multnomah_df['FEATURE_CLASS'] == 'Area'), :]
```

```python
multnomah_feature_df
```

```python
zipcodes = []
for index, row in multnomah_feature_df.iterrows():
    lat = row['PRIM_LAT_DEC']
    lng = row['PRIM_LONG_DEC']
    
    response = requests.get(f"{maps_base_url}latlng={lat},{lng}&key={gkey}").json()
#     print(len(response['results'][0]['address_components']))
    if response['results'][0]['address_components'][-1]['types'][0] == 'postal_code':
        zipcodes.append(response['results'][0]['address_components'][-1]['long_name'])
    elif response['results'][0]['address_components'][-2]['types'][0] == 'postal_code':
        zipcodes.append(response['results'][0]['address_components'][-2]['long_name'])
    else:
        zipcodes.append(response['results'][0]['address_components'][-3]['long_name'])
zipcodes
```

```python
multnomah_feature_df['Zipcode'] = zipcodes
multnomah_feature_df.columns
```

```python
multnomah_feature_df
mf_df_cleaned = multnomah_feature_df.iloc[:,[0,1,9,10,20]]
mf_df_cleaned.reset_index(inplace=True)
```

```python
mf_df_cleaned = mf_df_cleaned.drop(columns='index')
```
